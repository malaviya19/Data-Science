```python
c1 = np.array([-1,1])
b1 = np.random.randn(1000,2)*0.5 + c1
d1 = np.linalg.norm(b1-c1,axis=1)

c2 = np.array([-1,1])
b2 = np.random.randn(1000,2)*0.5 + c2
d2 = np.linalg.norm(b2-c2,axis=1)

cm = pl.get_cmap('jet')
pl.scatter(b1[:,0],b1[:,1], cmap = cm)
pl.scatter(b2[:,0], b2[:,1], cmap = cm)
pl.grid(True);
```

```python
import numpy as np

x1 = np.linspace(-5,5,100)
y1 = np.sin(x1)

x2 = np.array([-5,5])
y2 = np.array([0,0])

x3 = np.array([2,2])
y3 = np.array([-1,1])

x4 = np.pi
y4 = 0

x5 = 2.0
y5 = np.sin(x5)
```

```python
import matplotlib.pyplot as pl

pl.plot(x1,y1);
pl.plot(x2,y2, color='Gray')
pl.plot(x3,y3,color='Gray')
pl.scatter([x4,x5], [y4,y5], s=100)
```

```python
x = np.linspace(-10,10,100)
ySin = np.sin(x)
yCos = np.cos(x)
pl.plot(x,ySin)
pl.plot(x,yCos)
```

```python
pl.plot(x,ySin)
pl.plot(x,yCos)
pl.fill_between(x,ySin,yCos, where=ySin < yCos)
```
